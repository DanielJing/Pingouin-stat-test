```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.7, 1.2, 1.9, 2.7, 3.7, 4.3, 5.1, 6.4, 7.3, 8.7, 9.1, 12])
y = np.array([0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 12])
```

```python
def plot_data(x, y):
    plt.scatter(x, y)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.grid()

plt.show(plot_data(x, y))
```

```python
from sklearn.linear_model import SGDRegressor

X = x.reshape(-1, 1) #transpose single dimention data

reg = SGDRegressor(loss='squared_error')
reg.fit(X, y)

reg2 = SGDRegressor(loss='huber')
reg2.fit(X, y)
```

```python
def plot_regression_line(x, y, w0, w1, color, label):
    p_x = np.array([x.min(), x.max()])
    p_y = w0 + w1 * p_x
    plt.plot(p_x, p_y, color, label=label)

plot_data(x, y)
plot_regression_line(x, y, reg.intercept_, reg.coef_[0], 'r', label='Squared loss')
plot_regression_line(x, y, reg2.intercept_, reg2.coef_[0], 'm', label='Huber loss')
plt.legend()
plt.show()
```
